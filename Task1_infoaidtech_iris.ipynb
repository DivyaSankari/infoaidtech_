```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python

data1 = pd.read_csv("/content/IRIS.csv")
data1
```

```python

X = data1.drop("species", axis=1)
y = data1["species"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

sns.scatterplot(data=data1, x="sepal_length", y="sepal_width", hue="species")
plt.title("Scatterplot: Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
sns.scatterplot(data=data1, x="petal_length", y="petal_width", hue="species")
plt.title("Scatterplot: Petal Length vs. Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.show()
```

```python
sns.histplot(data=data1, x="sepal_length", hue="species", kde=True)
sns.histplot(data=data1, x="sepal_width", hue="species", kde=True)
sns.histplot(data=data1, x="petal_length", hue="species", kde=True)
sns.histplot(data=data1, x="petal_width", hue="species", kde=True)
plt.show()
```

```python

knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python

confusion_mat = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", confusion_mat)
```

```python

class_report = classification_report(y_test, y_pred)
print("Classification Report:\n", class_report)
```
